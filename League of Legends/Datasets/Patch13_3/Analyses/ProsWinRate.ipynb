Red Side vs Blue Side win rate

```python
wins_df = temp_df[temp_df["victory"] == True]
```

```python
b = (blueVictory[1]/ wins_df.shape[0]) * 100
b
```

```python
r = (redVictory[1]/ wins_df.shape[0]) * 100
r
```

---

Win rate for each player in each side

```python
players_df = np.unique(wins_df["summoner"])
games_df = np.unique(wins_df["gameId"])
for player in players_df:
    temp2_df = wins_df[wins_df["summoner"] == player]
    print("{} has a total of {} victories\n".format(player, temp2_df.shape[0]))


    #Player Red side Victories
    playerWinsRedSide_df = temp2_df[temp2_df["team"] == "Red"]
    playerWinsRedSide = len(np.unique(playerWinsRedSide_df["gameId"]))
    print("{} has {} victories on red side".format(player, playerWinsRedSide))

    #Player Red side Victories
    playerWinsBlueSide_df = temp2_df[temp2_df["team"] == "Blue"]
    playerWinsBlueSide = len(np.unique(playerWinsBlueSide_df["gameId"]))
    print("{} has {} victories on blue side\n".format(player, playerWinsBlueSide))

    print("{} has {:.2f} win rate on blue side".format(player, (playerWinsBlueSide/temp2_df.shape[0]) * 100))
    print("{} has {:.2f} win rate on red side\n".format(player, (playerWinsRedSide/temp2_df.shape[0]) * 100))

    print("-------------------------------------------\n")
```
