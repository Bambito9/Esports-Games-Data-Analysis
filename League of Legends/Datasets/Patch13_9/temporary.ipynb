### Professional Players Data Set

```python
df = pd.read_csv("Professional/FinalProfessionalPlayersDataset.csv")
df
```

```python
len(np.unique(df["gameId"]))
```

Remove remaked matches

```python
games = df[df["timestamp"] > 5]

remaked = []
totalMatches = len(np.unique(games["gameId"]))
count = 0
for match in np.unique(df["gameId"]):
    count += 1
    print("{} of {}".format(count, totalMatches))
    if((match in np.unique(games["gameId"])) == False):
        remaked.append(match)

print("Number of remaked matches: {}".format(len(remaked)))
```

```python
remaked
```

```python
for match in remaked:
    df.drop(df[df['gameId'] == match].index, inplace = True)
```

```python
len(np.unique(df["gameId"]))
```

```python
df.to_csv('Professional/FinalProfessionalPlayersDataset2.csv', index = False)
```
